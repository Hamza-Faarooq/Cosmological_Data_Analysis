```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Set random seed for reproducibility
np.random.seed(42)

# Step 1: Generate Synthetic Data
num_galaxies = 100000

# Generate synthetic data for galaxy properties
data = pd.DataFrame({
    'redshift': np.random.normal(loc=1.5, scale=0.5, size=num_galaxies),
    'mass': np.random.lognormal(mean=10, sigma=0.5, size=num_galaxies),
    'luminosity': np.random.lognormal(mean=10, sigma=0.5, size=num_galaxies),
    'temperature': np.random.normal(loc=2.7, scale=0.02, size=num_galaxies),
    'velocity': np.random.normal(loc=300, scale=100, size=num_galaxies),
    'age': np.random.normal(loc=13.8, scale=1, size=num_galaxies),
    'brightness': np.random.normal(loc=100, scale=20, size=num_galaxies)
})

# Step 2: Data Preprocessing
data = data[(data['redshift'] > 0) & (data['temperature'] > 0)]
data['log_mass'] = np.log(data['mass'])
data['log_luminosity'] = np.log(data['luminosity'])

# Step 3: Exploratory Data Analysis (EDA) & Step 4: Graphs and Plots

# Plot 1: Redshift Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['redshift'], bins=50, kde=True, color='skyblue')
plt.title('Redshift Distribution')
plt.xlabel('Redshift')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plot 2: Mass vs Luminosity Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='mass', y='luminosity', data=data, alpha=0.5)
plt.title('Mass vs Luminosity')
plt.xlabel('Mass')
plt.ylabel('Luminosity')
plt.grid(True)
plt.show()

# Plot 3: Temperature Distribution (CMB)
plt.figure(figsize=(10, 6))
sns.histplot(data['temperature'], bins=50, kde=True, color='orange')
plt.title('CMB Temperature Distribution')
plt.xlabel('Temperature (K)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plot 4: Velocity Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['velocity'], bins=50, kde=True, color='purple')
plt.title('Galaxy Velocity Distribution')
plt.xlabel('Velocity (km/s)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plot 5: Age of Galaxies
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=50, kde=True, color='green')
plt.title('Age of Galaxies')
plt.xlabel('Age (billion years)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plot 6: Brightness Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['brightness'], bins=50, kde=True, color='red')
plt.title('Brightness Distribution')
plt.xlabel('Brightness')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plot 7: Mass vs Redshift Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='redshift', y='mass', data=data, alpha=0.5, color='blue')
plt.title('Mass vs Redshift')
plt.xlabel('Redshift')
plt.ylabel('Mass')
plt.grid(True)
plt.show()

# Plot 8: Luminosity vs Redshift Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='redshift', y='luminosity', data=data, alpha=0.5, color='magenta')
plt.title('Luminosity vs Redshift')
plt.xlabel('Redshift')
plt.ylabel('Luminosity')
plt.grid(True)
plt.show()

# Plot 9: Redshift vs Temperature Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='redshift', y='temperature', data=data, alpha=0.5, color='cyan')
plt.title('Redshift vs Temperature')
plt.xlabel('Redshift')
plt.ylabel('Temperature (K)')
plt.grid(True)
plt.show()

# Plot 10: Correlation Matrix Heatmap
plt.figure(figsize=(10, 8))
corr = data[['redshift', 'mass', 'luminosity', 'temperature', 'velocity', 'age', 'brightness', 'log_mass', 'log_luminosity']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()

# Save the final DataFrame to a CSV file for further analysis
data.to_csv('synthetic_cosmology_data.csv', index=False)

print("Data generation, preprocessing, and visualization complete. The data has been saved to 'synthetic_cosmology_data.csv'.")
```
